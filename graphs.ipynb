```python
df = pd.read_csv('task3_results.csv')

plt.figure(figsize=(12, 6))
bars = plt.bar(df["month"], df["active_users"], color='skyblue')

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, height, f'{height}', ha='center', va='bottom')

plt.xlabel("Month")
plt.ylabel("Active Users")
plt.title("Active Users by Month")
plt.xticks(rotation=45, ha="right")
plt.tight_layout() 
plt.show()
```

```python
df = pd.read_csv('task3_results.csv')

plt.figure(figsize=(12, 6))
plt.plot(df["month"], df["running_sum_revenue"], marker='o', linestyle='-', color='purple')

for i, txt in enumerate(df["running_sum_revenue"]):
    plt.annotate(txt, (df["month"][i], df["running_sum_revenue"][i]), textcoords="offset points", xytext=(0,10), ha='center')

plt.xlabel("Month")
plt.ylabel("Cumulative Revenue")
plt.title("Cumulative Revenue by Month")
plt.xticks(rotation=45, ha="right")
plt.grid(True)
plt.tight_layout()

plt.show()
```
